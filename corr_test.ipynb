## Pearsonâ€™s Correlation Coefficient
Tests whether two samples have a linear relationship.

Assumptions

* Observations in each sample are independent and identically distributed (iid).
* Observations in each sample are normally distributed.
* Observations in each sample have the same variance.

Interpretation

* H0: the two samples are independent.
* H1: there is a dependency between the samples.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# this is just some data I collected myself from a machine learning models stock predictions
data = pd.read_csv('allonemonth_Mar27_Apr17.csv')
signal = data['signal']
percent_return = data['percent_return']
```

```python
# Example of the Pearson's Correlation test
from scipy.stats import pearsonr
stat, p = pearsonr(signal, percent_return)
print('stat=%.3f, p=%.5f' % (stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

p-value is low, so the null gets rejected
and we can conclude the two are probably correlated

```python
data.sort_values(by = 'signal', ascending = True).plot(x = 'signal', y = 'percent_return', 
                                                         figsize = (16,8), 
                                                         title='stock predictions Mar 27-Apr 17',
                                                             style = ' .')
plt.show()
```

```python
corr, _ = pearsonr(signal, percent_return)
print('Pearsons correlation: %.3f' % corr)
```

As we can see from the graph and from the correlation value there is a 
fairly weak positive correlation between the two values
